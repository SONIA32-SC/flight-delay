```python
#Importing libraries

import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# loading dataset

fl_process = pd.read_csv('fl_eda.csv')
```

```python
fl_process.head()
```

```python
# Removing unnecessary columns

fl_process = fl_process.drop(columns=['Unnamed: 0','AIRLINE_DELAY','DEPARTURE_DELAY','NAS_DELAY','DESTINATION_AIRPORT','ORIGIN_AIRPORT','LATE_AIRCRAFT_DELAY','iso_region','SECURITY_DELAY','id','ident','AIR_SYSTEM_DELAY','ARRIVAL_DELAY', 'WEATHER_DELAY','CANCELLATION_REASON', 'type','name','TAIL_NUMBER','latitude_deg','longitude_deg','elevation_ft','iso_country','iata_code'],axis= 1)
```

```python
fl_process.head()
```

```python
# Encoding categorical variables

fl_process = pd.get_dummies(fl_process, columns=['AIRLINE','IATA_CODE'])
fl_process
```

```python
# Choosing to store data types in smaller bits

fl_process[fl_process.select_dtypes(np.float64).columns] = fl_process.select_dtypes(np.float64).astype(np.float16)
```

```python
fl_process[fl_process.select_dtypes(np.int64).columns] = fl_process.select_dtypes(np.int64).astype(np.int16)
```

```python
# Defining X and y

X= fl_process.drop(columns=['CANCELLED', 'AIRLINES', 'IATA_CODE'])
y= fl_process['CANCELLED']
```

```python
# Splitting data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(fl_process, fl_process['CANCELLED'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Saving train and test sets

dataset_dict = {
    "X_train": X_train,
    "X_test": X_test,
    "y_train": y_train,
    "y_test": y_test
}

with open('dataset_dict.pickle', 'wb') as file:
    pickle.dump(dataset_dict, file)
```
